### Criptografía Clásica y Moderna

A. Cifrado César

```python
def cifrar_cesar(texto, clave):
    resultado = ""
    for char in texto:
        if char.isalpha():
            mayus = char.isupper()
            base = ord('A') if mayus else ord('a')
            resultado += chr((ord(char) - base + clave) % 26 + base)
        else:
            resultado += char
    return resultado

def descifrar_cesar(texto, clave):
    return cifrar_cesar(texto, -clave)

clave_cesar = 10

def fuerza_bruta_cesar(texto_cifrado):
    for k in range(26):
        print(f"Clave {k:2d}: {descifrar_cesar(texto_cifrado, k)}")

print("=== Cifrado César ===")
nombre = "Gabriela Espinosa"

cifrado = cifrar_cesar(nombre, clave_cesar)
print(f"Original: {nombre}")
print(f"Cifrado (clave={clave_cesar}): {cifrado}")
print(f"Descifrado: {descifrar_cesar(cifrado, clave_cesar)}")
print("\nAtaque de fuerza bruta sobre 'Jdeulhod Hvslqrvd':")
fuerza_bruta_cesar("Jdeulhod Hvslqrvd")
```

B. Cifrado Vigenère

```python
def cifrar_vigenere(texto, clave):
    texto = texto.upper() #todo en mayuscula
    clave = clave.upper()
    resultado = ""
    idx_clave = 20 #indice para recorrer la clave
    for char in texto:
        if char.isalpha():
            desplazamiento = ord(clave[idx_clave % len(clave)]) - ord('A')
            resultado += chr((ord(char) - ord('A') + desplazamiento) % 26 + ord('A'))
            idx_clave += 1
        else:
            resultado += char
    return resultado

def descifrar_vigenere(texto, clave):
    texto = texto.upper()
    clave = clave.upper()
    resultado = ""
    idx_clave = 20
    for char in texto:
        if char.isalpha():
            desplazamiento = ord(clave[idx_clave % len(clave)]) - ord('A')
            resultado += chr((ord(char) - ord('A') - desplazamiento) % 26 + ord('A'))
            idx_clave += 1
        else:
            resultado += char
    return resultado

print("\n=== Cifrado Vigenère ===")
mensaje = "El profe es super genial uwu"
clave_vigenere = "VIGNERE"
cifrado_vig = cifrar_vigenere(mensaje, clave_vigenere)
print(f"Original: {mensaje}")
print(f"Cifrado con clave '{clave_vigenere}': {cifrado_vig}")
print(f"Descifrado: {descifrar_vigenere(cifrado_vig, clave_vigenere)}")
```

Cifrado Simétrico Moderno (AES)

```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes
import binascii

def aes_ejemplo():
    print("\n=== Cifrado AES (Modos ECB y CBC) ===")

    # Generar una clave aleatoria de 128 bits (16 bytes)
    clave_secreta = get_random_bytes(16)

    # Mensaje de prueba
    texto_original = "No hay lugar como 127.0.0.1"
    datos_bytes = texto_original.encode('utf-8')

    # ---- MODO CBC ----
    iv = get_random_bytes(16)  # Vector de inicialización
    cifrador_cbc = AES.new(clave_secreta, AES.MODE_CBC, iv=iv)
    # El padding se añade para que el mensaje sea múltiplo del tamaño de bloque (16 bytes)
    cifrado_cbc = cifrador_cbc.encrypt(pad(datos_bytes, AES.block_size))

    print("\n[Modo CBC]")
    #print(f"IV (hex): {binascii.hexlify(iv).decode()}")
    print(f"Cifrado (hex): {binascii.hexlify(cifrado_cbc).decode()}")

    # Descifrado CBC
    descifrador_cbc = AES.new(clave_secreta, AES.MODE_CBC, iv=iv)
    datos_descifrados_cbc = unpad(descifrador_cbc.decrypt(cifrado_cbc), AES.block_size)
    print(f"Descifrado: {datos_descifrados_cbc.decode()}")

    # ---- MODO ECB ----
    cifrador_ecb = AES.new(clave_secreta, AES.MODE_ECB)
    cifrado_ecb = cifrador_ecb.encrypt(pad(datos_bytes, AES.block_size))

    print("\n[Modo ECB]")
    print(f"Cifrado (hex): {binascii.hexlify(cifrado_ecb).decode()}")

    # Descifrado ECB
    descifrador_ecb = AES.new(clave_secreta, AES.MODE_ECB)
    datos_descifrados_ecb = unpad(descifrador_ecb.decrypt(cifrado_ecb), AES.block_size)
    print(f"Descifrado: {datos_descifrados_ecb.decode()}")

    # ---- Cifrado de archivo ----
    print("\n--- Cifrado de archivo con AES CBC ---")
    nombre_archivo = "/content/Mensaje_secreto.txt"
    with open(nombre_archivo, "wb") as f:
        f.write(b"Give a man a fish, and he'll eat for a day. Teach a man phishing, and he'll spam you for the rest of his life.")

    # Leer archivo
    with open(nombre_archivo, "rb") as f:
        archivo_bytes = f.read()

    # Cifrar con nueva clave e IV
    clave_archivo = get_random_bytes(16)
    iv_archivo = get_random_bytes(16)
    cifrador_archivo = AES.new(clave_archivo, AES.MODE_CBC, iv=iv_archivo)
    archivo_padded = pad(archivo_bytes, AES.block_size)
    archivo_cifrado = cifrador_archivo.encrypt(archivo_padded)

    # Guardar archivo cifrado junto con el IV (para poder descifrar después)
    with open("/content/Descifrado.txt", "wb") as f:
        f.write(iv_archivo + archivo_cifrado)  # Guardamos IV + datos cifrados
    print("Archivo cifrado guardado como '/content/Descifrado.txt'")

    # Descifrar archivo
    with open("/content/Descifrado.txt", "rb") as f:
        iv_guardado = f.read(16)
        datos_cifrados = f.read()

    descifrador_archivo = AES.new(clave_archivo, AES.MODE_CBC, iv=iv_guardado)
    datos_descifrados = unpad(descifrador_archivo.decrypt(datos_cifrados), AES.block_size)
    print(f"Archivo descifrado correctamente: {datos_descifrados.decode()}")

# Llamar a la función
aes_ejemplo()
```

### Funciones Hash y HMAC

A. SHA-256

```python
import hashlib
def hash_texto(texto):
    return hashlib.sha256(texto.encode('utf-8')).hexdigest()

print("\n=== Hash SHA-Textos ===")
texto1 = "Esternocleidomastoideo"
texto2 = "Ezternocleidomastoideo"
texto3 = "Esternocleydomastoideo"
print(f"Hash de '{texto1}': {hash_texto(texto1)}")
print(f"Hash de '{texto2}': {hash_texto(texto2)}")
print(f"Hash de '{texto3}': {hash_texto(texto3)}")



def generar_archivo_hash(archivo_entrada, archivo_salida_hash=None):
    """Calcula el hash SHA-256 de un archivo y lo guarda en un archivo de texto."""
    if archivo_salida_hash is None:
        archivo_salida_hash = archivo_entrada + ".sha256"
    hash_val = hash_archivo(archivo_entrada)
    with open(archivo_salida_hash, 'w') as f:
        f.write(hash_val)
    print(f"Hash guardado en {archivo_salida_hash}")
    return hash_val


#Hash Archivo1
print("\n--- Generación y verificación de archivo de hash ---")
generar_archivo_hash("archivo1.txt")  # Crea archivo1.txt.sha256


#Hash Archivo2
print("\n--- Generación y verificación de archivo de hash ---")
generar_archivo_hash("archivo2.txt")  # Crea archivo2.txt.sha256

#Hash Archivo3
print("\n--- Generación y verificación de archivo de hash ---")
generar_archivo_hash("archivo3.txt")  # Crea archivo3.txt.sha256

```

B. Detección de integridad

```python
import hashlib
def hash_archivo(ruta):
    sha = hashlib.sha256()
    with open(ruta, 'rb') as f:
        for bloque in iter(lambda: f.read(4096), b""):
            sha.update(bloque)
    return sha.hexdigest()
# Comparar archivos
def comparar_archivos_hash(ruta1, ruta2):
    return hash_archivo(ruta1) == hash_archivo(ruta2)

with open("archivo1.txt", "w") as f:
    f.write("El profesor nos coloca cinco uwu")
with open("archivo2.txt", "w") as f:
    f.write("El profesor nos coloca cinco uwu")
with open("archivo3.txt", "w") as f:
    f.write("El profesor nos coloca tres unu")

print(f"\n¿archivo1.txt y archivo2.txt iguales? {comparar_archivos_hash('archivo1.txt', 'archivo2.txt')}")
print(f"¿archivo1.txt y archivo3.txt iguales? {comparar_archivos_hash('archivo1.txt', 'archivo3.txt')}")
print(f"¿archivo2.txt y archivo3.txt iguales? {comparar_archivos_hash('archivo2.txt', 'archivo3.txt')}")
```

C. HMAC

```python
import hmac
def generar_hmac(mensaje, clave_hmac):
    return hmac.new(clave_hmac.encode(), mensaje.encode(), hashlib.sha256).hexdigest()

def verificar_hmac(mensaje, clave_hmac, hmac_recibido):
    calculado = generar_hmac(mensaje, clave_hmac)
    return hmac.compare_digest(calculado, hmac_recibido)

print("\n=== HMAC ===")
msg = "¿Cómo vendió el horticultor sus productos en la red oscura? Utilizó el enrutamiento de cebolla."
key_hmac = "Cebollas"
hmac_val = generar_hmac(msg, key_hmac)
print(f"HMAC calculado: {hmac_val}")
print(f"Verificación correcta: {verificar_hmac(msg, key_hmac, hmac_val)}")
print(f"Verificación con mensaje alterado: {verificar_hmac(msg + 'X', key_hmac, hmac_val)}")
```

### Autenticación con hash y sal

```python
import os

base_usuarios = {}

def registrar(usuario, password):
    sal = os.urandom(16).hex()
    hash_pw = hashlib.pbkdf2_hmac('sha256', password.encode(), sal.encode(), 100000).hex()
    base_usuarios[usuario] = {"hash": hash_pw, "sal": sal}
    print(f"Usuario '{usuario}' registrado con éxito.")

def login(usuario, password):
    if usuario not in base_usuarios:
        return False
    datos = base_usuarios[usuario]
    hash_calculado = hashlib.pbkdf2_hmac('sha256', password.encode(), datos["sal"].encode(), 100000).hex()
    return hmac.compare_digest(hash_calculado, datos["hash"])

while True:
  print("\n--- Menú de Autenticación ---")
  print("1. Registrar nuevo usuario")
  print("2. Iniciar sesión")
  print("3. Salir")
  opcion = input("Seleccione una opción: ").strip()
  if opcion == '1':
      user = input("Ingrese nombre de usuario: ").strip()
      if user in base_usuarios:
          print("El usuario ya existe. Intente con otro nombre.")
          continue
      pwd = input("Ingrese contraseña: ").strip()
      registrar(user, pwd)
  elif opcion == '2':
      user = input("Usuario: ").strip()
      pwd = input("Contraseña: ").strip()
      if login(user, pwd):
          print("Inicio de sesión exitoso. ¡Bienvenido!")
      else:
          print("Usuario o contraseña incorrectos.")
  elif opcion == '3':
      print("Saliendo del modo interactivo.")
      break
  else:
      print("Opción no válida.")
```

### Firma Digital

```python
from Crypto.PublicKey import RSA
from Crypto.Signature import pkcs1_15
from Crypto.Hash import SHA256


def firma_digital():
    """
    Demuestra la generación y verificación de una firma digital usando RSA.
    """
    print("\n=== Firma Digital RSA ===")

    # 1. Generar par de claves (pública y privada)
    key = RSA.generate(2048)          # Clave RSA de 2048 bits
    private_key = key.export_key()     # Formato PEM (para guardar)
    public_key = key.publickey().export_key()

    print("[OK] Par de claves RSA generado.")

    # 2. Mensaje a firmar
    mensaje = "¿Por qué la IA se negó a jugar al escondite? ¡Porque sabía que no podía ocultar sus violaciones de datos!".encode('utf-8')
    print(f"Mensaje original: {mensaje.decode('utf-8')}") # Decode for readable printing

    # 3. Calcular el hash del mensaje (SHA-256)
    h = SHA256.new(mensaje)

    # 4. Firmar el hash con la clave privada
    firmador = pkcs1_15.new(key)
    firma = firmador.sign(h)
    print(f"Firma generada (hex): {firma.hex()[:64]}... ")

    # 5. Verificar la firma con la clave pública
    verificador = pkcs1_15.new(key.publickey())
    try:
        verificador.verify(h, firma)
        print("Verificación exitosa: la firma es válida.")
    except (ValueError, TypeError):
        print("Firma inválida.")

    # 6. Demostrar que si el mensaje cambia, la verificación fallo
    mensaje_modificado = "¿Por qué la IA se negó a jugar al escondite? no se".encode('utf-8')
    print(f"Mensaje modificado {mensaje_modificado.decode('utf-8')}" )
    h_mod = SHA256.new(mensaje_modificado)
    try:
        verificador.verify(h_mod, firma)
        print("La verificación con mensaje modificado pasa.")
    except (ValueError, TypeError):
        print("Verificación fallida: el mensaje fue alterado.")

    #Guardar las claves en archivos para uso posterior
    with open("clave_privada.pem", "wb") as f:
        f.write(private_key)
    with open("clave_publica.pem", "wb") as f:
        f.write(public_key)
    print("\nClaves guardadas en 'clave_privada.pem' y 'clave_publica.pem'.")

# Llamar a la función
if __name__ == "__main__":
    firma_digital()
```

### Introducción Práctica a Blockchain

```python
import hashlib
import datetime

class BloqueEstudiante:
    """
    Bloque que almacena la información de un estudiante.
    """
    def __init__(self, nombre, edad, codigo, genero, hash_anterior=""):
        self.timestamp = datetime.datetime.now().isoformat()
        self.nombre = nombre
        self.edad = edad
        self.codigo = codigo
        self.genero = genero
        self.hash_anterior = hash_anterior
        self.nonce = 0
        self.hash = self.calcular_hash()

    def calcular_hash(self):
        """
        Calcula el hash SHA-256 del contenido del bloque.
        """
        contenido = (str(self.timestamp) + self.nombre + str(self.edad) +
                     self.codigo + self.genero + self.hash_anterior + str(self.nonce))
        return hashlib.sha256(contenido.encode()).hexdigest()

    def minar_bloque(self, dificultad):
        """
        Prueba de trabajo: busca un nonce que haga que el hash comience con 'dificultad' ceros.
        """
        objetivo = "0" * dificultad
        while self.hash[:dificultad] != objetivo:
            self.nonce += 1
            self.hash = self.calcular_hash()
        print(f"Bloque minado para {self.nombre}: {self.hash[:10]}... (nonce={self.nonce})")

    def __repr__(self):
        return (f"BloqueEstudiante(\n"
                f"  timestamp    : {self.timestamp}\n"
                f"  nombre       : {self.nombre}\n"
                f"  edad         : {self.edad}\n"
                f"  código       : {self.codigo}\n"
                f"  género       : {self.genero}\n"
                f"  hash_anterior: {self.hash_anterior[:10]}...\n"
                f"  hash         : {self.hash[:10]}...\n"
                f"  nonce        : {self.nonce}\n)")


class BlockchainEstudiantil:
    """
    Cadena de bloques para el registro de estudiantes.
    """
    def __init__(self, dificultad=4):
        self.cadena = [self.crear_bloque_maria()]
        self.dificultad = dificultad

    def crear_bloque_maria(self):
        """
        Bloque inicial.
        """
        return BloqueEstudiante(
            nombre="Maria Fer",
            edad=23,
            codigo="1202412",
            genero="F",
            hash_anterior="0"
        )

    def obtener_ultimo_bloque(self):
        return self.cadena[-1]

    def agregar_estudiante(self, nombre, edad, codigo, genero):
        """
        Crea un nuevo bloque con los datos del estudiante y lo añade a la cadena.
        """
        nuevo = BloqueEstudiante(
            nombre, edad, codigo, genero,
            hash_anterior=self.obtener_ultimo_bloque().hash
        )
        nuevo.minar_bloque(self.dificultad)
        self.cadena.append(nuevo)
        print(f"Estudiante registrado: {nombre} (código {codigo})")

    def es_cadena_valida(self):
        """
        Verifica la integridad de toda la cadena.
        """
        for i in range(1, len(self.cadena)):
            actual = self.cadena[i]
            anterior = self.cadena[i-1]

            # Recalcular el hash del bloque actual
            if actual.hash != actual.calcular_hash():
                print(f"ERROR: El hash del bloque de {actual.nombre} no coincide.")
                return False

            # Verificar el enlace con el bloque anterior
            if actual.hash_anterior != anterior.hash:
                print(f"ERROR: El bloque de {actual.nombre} no apunta correctamente al anterior.")
                return False

        return True

    def mostrar_cadena(self):
        """Imprime todos los bloques de la cadena."""
        print("\n=== CADENA DE BLOQUES ESTUDIANTIL ===\n")
        for i, bloque in enumerate(self.cadena):
            print(f"--- Bloque {i} ---")
            print(bloque)
            print()


# =============================================================================
# DEMOSTRACIÓN
# =============================================================================
if __name__ == "__main__":
    # Crear la blockchain con dificultad baja (3 ceros) para que sea rápida
    blockchain = BlockchainEstudiantil(dificultad=3)

    # Agregar algunos estudiantes (puedes reemplazar con tus datos reales)
    print("\n--- Registrando estudiantes en la blockchain ---\n")
    blockchain.agregar_estudiante("Gabriela", 21, "1202419", "F")
    blockchain.agregar_estudiante("Gian Marco", 22, "1202414", "M")
    blockchain.agregar_estudiante("Michelle", 20, "362514", "F")
    blockchain.agregar_estudiante("Pedro Sánchez", 23, "202311", "M")

    # Mostrar toda la cadena
    blockchain.mostrar_cadena()

    # Verificar integridad
    print("Verificando integridad de la cadena...")
    print("Cadena válida:", blockchain.es_cadena_valida())

    # --- Demostración de inmutabilidad ---
    print("\n--- Modificando un bloque (ataque) ---")
    # Accedemos al segundo bloque (índice 1) y cambiamos la edad
    bloque_modificado = blockchain.cadena[1]
    print(f"Bloque original de {bloque_modificado.nombre}: edad={bloque_modificado.edad}")
    bloque_modificado.edad = 99  # Modificación fraudulenta
    # Para que el bloque sea válido, el atacante debería recalcular su hash (pero no puede cambiar los siguientes)
    bloque_modificado.hash = bloque_modificado.calcular_hash()  # El atacante re-mina este bloque
    print(f"Bloque modificado: ahora edad={bloque_modificado.edad}")

    print("\nVerificando cadena después de la modificación...")
    if not blockchain.es_cadena_valida():
        print("La cadena ha sido alterada y ya no es válida.")
    else:
        print("La cadena sigue siendo válida.")

    # Explicación: aunque el atacante recalcule el hash del bloque modificado,
    # los bloques siguientes (índice 2 en adelante) aún apuntan al hash anterior original,
    # por lo que la verificación fallará (a menos que se re-minen todos los bloques siguientes).
```

### Laboratorio sobre HTTPS

```python
import urllib.request
import urllib.error
import ssl
import socket

def prueba_https():
    print("=== Laboratorio HTTPS ===\n")

    # 1. Sitio con HTTPS
    url_https = "https://www.tablesgenerator.com/"
    print(f"[HTTPS] Solicitando {url_https} ...")
    try:
        # Añadimos un User-Agent para evitar bloqueos
        req = urllib.request.Request(url_https, headers={'User-Agent': 'Mozilla/5.0'})
        with urllib.request.urlopen(req, timeout=5) as respuesta:
            print(f"  Código de respuesta: {respuesta.status} {respuesta.reason}")
            print("  Encabezados relevantes:")
            for header, value in respuesta.getheaders():
                if header.lower() in ('content-type', 'server', 'date'):
                    print(f"    {header}: {value}")
            # Leemos solo los primeros bytes para no saturar
            contenido = respuesta.read(200)
            print(f"  Primeros 200 bytes del contenido: {contenido[:100]}...\n")
    except urllib.error.URLError as e:
        print(f"  Error en petición HTTPS: {e}\n")

    # 2. Sitio sin HTTPS (solo HTTP)
    url_http = "http://httpbin.org/"
    print(f"[HTTP] Solicitando {url_http} ...")
    try:
        req = urllib.request.Request(url_http, headers={'User-Agent': 'Mozilla/5.0'})
        with urllib.request.urlopen(req, timeout=5) as respuesta:
            print(f"  Código de respuesta: {respuesta.status} {respuesta.reason}")
            print("  Encabezados relevantes:")
            for header, value in respuesta.getheaders():
                if header.lower() in ('content-type', 'server', 'date'):
                    print(f"    {header}: {value}")
            print("  (Conexión sin cifrar, los datos viajan en texto claro)\n")
    except urllib.error.URLError as e:
        print(f"  Error en petición HTTP: {e}\n")

    # 3. Verificación de certificado expirado
    url_expirado = "https://expired.badssl.com/"
    print(f"[Certificado expirado] Solicitando {url_expirado} ...")
    try:
        with urllib.request.urlopen(url_expirado, timeout=5) as respuesta:
            print(" El certificado es inválido.\n")
    except urllib.error.URLError as e:
        if isinstance(e.reason, ssl.SSLCertVerificationError):
            print("  Error: certificado expirado detectado.")
            print(f"  Detalle: {e.reason}\n")
        else:
            print(f"  Otro error: {e}\n")


if __name__ == "__main__":
    prueba_https()
```
