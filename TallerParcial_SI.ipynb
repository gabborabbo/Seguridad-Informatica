A. Cifrado César

```python
# Definimos una función llamada cifrar_cesar que recibe un texto y una clave (el número de posiciones a desplazar)
def cifrar_cesar(texto, clave):
    resultado = "" # Aquí iremos guardando el texto cifrado, letra por letra
    # Recorremos cada carácter del texto original
    for char in texto:
      # Preguntamos si el carácter es una letra (del alfabeto)
        if char.isalpha():
           # Averiguamos si la letra es mayúscula
            mayus = char.isupper()
            # Elegimos la base: si es mayúscula usamos el código ASCII de 'A', si no, el de 'a'
            base = ord('A') if mayus else ord('a')
            # Para cifrar:
            # 1. Convertimos la letra a número (ord) y le restamos la base para obtener un número entre 0 y 25
            # 2. Sumamos la clave (el desplazamiento) y aplicamos módulo 26 para que si se pasa de la 'Z' vuelva a empezar
            # 3. Volvemos a convertir ese número a letra sumándole la base
            resultado += chr((ord(char) - base + clave) % 26 + base)
        else:
          # Si no es una letra (espacio, coma, etc.), lo dejamos igual
            resultado += char
    # Devolvemos el texto ya cifrado
    return resultado

# Llamamos a cifrar pero con la clave negativa
def descifrar_cesar(texto, clave):
    return cifrar_cesar(texto, -clave)

# Elegimos una clave para nuestro ejemplo (desplazaremos 10 posiciones)
clave_cesar = 10

# Probamos todas las claves posibles (de 0 a 25)
def fuerza_bruta_cesar(texto_cifrado):
  # Recorremos todas las claves
    for k in range(26):
      # Mostramos la clave y el texto descifrado con esa clave
        print(f"Clave {k:2d}: {descifrar_cesar(texto_cifrado, k)}")

print("=== Cifrado César Nombre===")
# Definimos un nombre de ejemplo
nombre = "Gabriela Espinosa"

# Ciframos el nombre con la clave que elegimos
cifrado = cifrar_cesar(nombre, clave_cesar)
# Imprimimos el original, el cifrado y volvemos a descifrar para verificar que todo funciona
print(f"Original: {nombre}")
print(f"Cifrado (clave={clave_cesar}): {cifrado}")
print(f"Descifrado: {descifrar_cesar(cifrado, clave_cesar)}")
# Ahora probamos el ataque de fuerza bruta
print("\nAtaque de fuerza bruta sobre 'Qklbsovk Oczsxyck':")
fuerza_bruta_cesar("Qklbsovk Oczsxyck")

# Lo mismo pero con la frase
print("\n=== Cifrado César Frase===")

Frase = "El tiempo es lo que determina la seguridad. Con tiempo suficiente nada es inhackeable"

cifrado = cifrar_cesar(Frase, clave_cesar)
print(f"Original: {Frase}")
print(f"Cifrado (clave={clave_cesar}): {cifrado}")
print(f"Descifrado: {descifrar_cesar(cifrado, clave_cesar)}")

print("\nAtaque de fuerza bruta sobre 'Ov dsowzy oc vy aeo nodobwsxk vk coqebsnkn. Myx dsowzy cepsmsoxdo xknk oc sxrkmuoklvo':")
fuerza_bruta_cesar("Ov dsowzy oc vy aeo nodobwsxk vk coqebsnkn. Myx dsowzy cepsmsoxdo xknk oc sxrkmuoklvo")
```

B. Cifrado Vigenère

```python
# Función para cifrar un mensaje usando el método de Vigenère
def cifrar_vigenere(texto, clave):
    # Convertimos el texto a mayúsculas para trabajar solo con un tipo de letra
    texto = texto.upper()
    # También convertimos la clave a mayúsculas
    clave = clave.upper()
    # Variable donde iremos guardando el resultado cifrado
    resultado = ""
    # Índice que nos ayudará a recorrer la clave letra por letra
    idx_clave = 0 # Empieza en 0 para usar la primera letra de la clave
    # Recorremos cada carácter del texto original
    for char in texto:
        # Verificamos si el carácter es una letra del alfabeto
        if char.isalpha():
            # Calculamos el desplazamiento que corresponde a la letra actual de la clave
            # Restamos el código ASCII de 'A' para obtener un número entre 0 y 25
            desplazamiento = ord(clave[idx_clave % len(clave)]) - ord('A')
            # Para cifrar la letra:
            # 1. Convertimos la letra del texto a número (restando 'A')
            # 2. Sumamos el desplazamiento calculado
            # 3. Aplicamos módulo 26 para que el resultado se mantenga entre 0 y 25
            # 4. Volvemos a convertir el número a letra sumando 'A'
            resultado += chr((ord(char) - ord('A') + desplazamiento) % 26 + ord('A'))
            # Avanzamos al siguiente carácter de la clave (solo cuando ya ciframos una letra)
            idx_clave += 1
        else:
            # Si no es una letra, lo dejamos igual
            resultado += char
    # Devolvemos el texto ya cifrado
    return resultado

# Función para descifrar un mensaje cifrado con Vigenère
def descifrar_vigenere(texto, clave):
    # Convertimos todo a mayúsculas igual que en el cifrado
    texto = texto.upper()
    clave = clave.upper()
    resultado = ""
    # Iniciamos el índice de la clave en 0
    idx_clave = 0
    # Recorremos cada carácter del texto cifrado
    for char in texto:
        if char.isalpha():
            # Calculamos el desplazamiento de la clave igual que antes
            desplazamiento = ord(clave[idx_clave % len(clave)]) - ord('A')
            # Para descifrar, en lugar de sumar el desplazamiento, lo restamos
            resultado += chr((ord(char) - ord('A') - desplazamiento) % 26 + ord('A'))
            idx_clave += 1
        else:
            resultado += char
    return resultado

# Probamos el cifrado con un mensaje y una clave
print("\n=== Cifrado Vigenère ===")
mensaje = "El profe es super genial uwu"          # Mensaje original
clave_vigenere = "VIGENERE"                         # Clave que usaremos
cifrado_vig = cifrar_vigenere(mensaje, clave_vigenere)  # Ciframos el mensaje
print(f"Original: {mensaje}")
print(f"Cifrado con clave '{clave_vigenere}': {cifrado_vig}")
# Desciframos para comprobar que recuperamos el original (en mayúsculas, porque así lo convertimos)
print(f"Descifrado: {descifrar_vigenere(cifrado_vig, clave_vigenere)}")
```

Cifrado Simétrico Moderno (AES)

```python
# Importamos las herramientas necesarias de la librería pycryptodome
from Crypto.Cipher import AES          # Para usar el algoritmo AES
from Crypto.Util.Padding import pad, unpad  # Para ajustar los mensajes al tamaño correcto
from Crypto.Random import get_random_bytes  # Para generar bytes aleatorios (claves, IVs)
import binascii  # Para convertir los datos cifrados a hexadecimal y poder verlos

# Definimos una función que contiene todo el ejemplo de AES
def aes_ejemplo():
    print("\n=== Cifrado AES (Modos ECB y CBC) ===")

    # Generamos una clave secreta aleatoria de 16 bytes (128 bits)
    # Esta clave la deben conocer el que cifra y el que descifra
    clave_secreta = get_random_bytes(16)

    # El mensaje que vamos a cifrar
    texto_original = "No hay lugar como 127.0.0.1"
    # Convertimos el texto a bytes porque AES trabaja con bytes, no con texto directamente
    datos_bytes = texto_original.encode('utf-8')

    # ---- MODO CBC ----
    # En el modo CBC necesitamos un vector de inicialización (IV) aleatorio
    iv = get_random_bytes(16)  # El IV debe ser del mismo tamaño que el bloque (16 bytes)
    # Creamos un objeto "cifrador" en modo CBC, usando la clave y el IV
    cifrador_cbc = AES.new(clave_secreta, AES.MODE_CBC, iv=iv)

    # Antes de cifrar, tenemos que ajustar el mensaje para que su longitud sea múltiplo de 16 bytes.
    # Para eso usamos pad() que añade bytes de relleno.
    cifrado_cbc = cifrador_cbc.encrypt(pad(datos_bytes, AES.block_size))

    print("\n[Modo CBC]")
    # Mostramos el resultado cifrado en hexadecimal para que sea legible
    print(f"Cifrado (hex): {binascii.hexlify(cifrado_cbc).decode()}")

    # Para descifrar, creamos un descifrador con la misma clave y el mismo IV
    descifrador_cbc = AES.new(clave_secreta, AES.MODE_CBC, iv=iv)
    # Desciframos y luego quitamos el relleno con unpad()
    datos_descifrados_cbc = unpad(descifrador_cbc.decrypt(cifrado_cbc), AES.block_size)
    # Convertimos los bytes de vuelta a texto y lo mostramos
    print(f"Descifrado: {datos_descifrados_cbc.decode()}")

    # ---- MODO ECB ----
    # El modo ECB no necesita IV, porque cada bloque se cifra independientemente
    cifrador_ecb = AES.new(clave_secreta, AES.MODE_ECB)
    # Ciframos (también con padding)
    cifrado_ecb = cifrador_ecb.encrypt(pad(datos_bytes, AES.block_size))

    print("\n[Modo ECB]")
    print(f"Cifrado (hex): {binascii.hexlify(cifrado_ecb).decode()}")

    # Descifrado ECB
    descifrador_ecb = AES.new(clave_secreta, AES.MODE_ECB)
    datos_descifrados_ecb = unpad(descifrador_ecb.decrypt(cifrado_ecb), AES.block_size)
    print(f"Descifrado: {datos_descifrados_ecb.decode()}")

    # ---- Cifrado de archivo ----
    print("\n--- Cifrado de archivo con AES CBC ---")
    # Vamos a crear un archivo de texto con un mensaje
    nombre_archivo = "/content/Mensaje_secreto.txt"
    with open(nombre_archivo, "wb") as f:
        f.write(b"Give a man a fish, and he'll eat for a day. Teach a man phishing, and he'll spam you for the rest of his life.")

    # Leemos el contenido del archivo (en bytes)
    with open(nombre_archivo, "rb") as f:
        archivo_bytes = f.read()

    # Para el archivo usamos una nueva clave y un nuevo IV (para no mezclar con el ejemplo anterior)
    clave_archivo = get_random_bytes(16)
    iv_archivo = get_random_bytes(16)
    cifrador_archivo = AES.new(clave_archivo, AES.MODE_CBC, iv=iv_archivo)
    # Añadimos padding y ciframos
    archivo_padded = pad(archivo_bytes, AES.block_size)
    archivo_cifrado = cifrador_archivo.encrypt(archivo_padded)

    # Guardamos el archivo cifrado. Primero escribimos el IV (para poder recuperarlo al descifrar)
    # y luego los datos cifrados.
    with open("/content/Descifrado.txt", "wb") as f:
        f.write(iv_archivo + archivo_cifrado)  # Guardamos IV + datos cifrados
    print("Archivo cifrado guardado como '/content/Descifrado.txt'")

    # Ahora desciframos el archivo que acabamos de guardar
    with open("/content/Descifrado.txt", "rb") as f:
        iv_guardado = f.read(16)          # Leemos los primeros 16 bytes que son el IV
        datos_cifrados = f.read()          # El resto son los datos cifrados

    # Creamos un descifrador con la misma clave y el IV que recuperamos
    descifrador_archivo = AES.new(clave_archivo, AES.MODE_CBC, iv=iv_guardado)
    # Desciframos y quitamos el padding
    datos_descifrados = unpad(descifrador_archivo.decrypt(datos_cifrados), AES.block_size)
    # Mostramos el contenido original (debería ser el mismo que escribimos)
    print(f"Archivo descifrado correctamente: {datos_descifrados.decode()}")

# Llamamos a la función para que se ejecute todo
aes_ejemplo()
```

### Funciones Hash y HMAC

A. SHA-256

```python
import hashlib  # Importamos la librería que nos permite usar funciones hash

# Función para calcular el hash SHA-256 de un texto
def hash_texto(texto):
    # Convertimos el texto a bytes porque hashlib trabaja con bytes, no con texto directamente
    # .encode('utf-8') transforma el texto en bytes
    # .hexdigest() nos da el resultado en formato hexadecimal (letras y números)
    return hashlib.sha256(texto.encode('utf-8')).hexdigest()

# Función para calcular el hash SHA-256 de un archivo
def hash_archivo(ruta):
    # Creamos un objeto hash (vacío)
    sha = hashlib.sha256()
    # Abrimos el archivo en modo lectura de bytes ('rb')
    with open(ruta, 'rb') as f:
        # Leemos el archivo en bloques de 4096 bytes para no saturar la memoria si es muy grande
        # iter(lambda: f.read(4096), b"") es una forma de leer hasta que no haya más datos
        for bloque in iter(lambda: f.read(4096), b""):
            # Vamos actualizando el hash con cada bloque leído
            sha.update(bloque)
    # Devolvemos el hash final en hexadecimal
    return sha.hexdigest()

print("\n=== Hash SHA-256 de Textos ===")
# Definimos tres palabras muy parecidas, solo cambian una letra
texto1 = "Esternocleidomastoideo"
texto2 = "Ezternocleidomastoideo"  # Cambiamos la 's' por 'z'
texto3 = "Esternocleydomastoideo"  # Cambiamos la 'i' por 'y'

# Mostramos los hashes de cada palabra para ver cómo un pequeño cambio altera completamente el resultado
print(f"Hash de '{texto1}': {hash_texto(texto1)}")
print(f"Hash de '{texto2}': {hash_texto(texto2)}")
print(f"Hash de '{texto3}': {hash_texto(texto3)}")

# Función para guardar el hash de un archivo en otro archivo de texto
def generar_archivo_hash(archivo_entrada, archivo_salida_hash=None):
    """Calcula el hash SHA-256 de un archivo y lo guarda en un archivo de texto."""
    # Si no nos dan un nombre para el archivo de salida, lo creamos añadiendo ".sha256" al nombre original
    if archivo_salida_hash is None:
        archivo_salida_hash = archivo_entrada + ".sha256"
    # Calculamos el hash del archivo de entrada usando la función hash_archivo
    hash_val = hash_archivo(archivo_entrada)
    # Guardamos ese hash en un archivo de texto (en modo escritura)
    with open(archivo_salida_hash, 'w') as f:
        f.write(hash_val)
    print(f"Hash guardado en {archivo_salida_hash}")
    return hash_val

# Antes de generar los hashes, necesitamos que existan los archivos.
# Aquí los creamos con contenido específico.
with open("archivo1.txt", "w") as f:
    f.write("El profesor nos coloca cinco uwu")
with open("archivo2.txt", "w") as f:
    f.write("El profesor nos coloca cinco uwu")
with open("archivo3.txt", "w") as f:
    f.write("El profesor nos coloca tres unu")

# Archivo1.txt y archivo2.txt tienen el mismo contenido, por lo que sus hashes serán iguales.
# archivo3.txt tiene contenido diferente, así que su hash será distinto.

# Generamos el archivo de hash para archivo1.txt
print("\n--- Generación de archivo de hash para archivo1.txt ---")
generar_archivo_hash("archivo1.txt")  # Esto crea "archivo1.txt.sha256"

# Generamos el archivo de hash para archivo2.txt
print("\n--- Generación de archivo de hash para archivo2.txt ---")
generar_archivo_hash("archivo2.txt")  # Esto crea "archivo2.txt.sha256"

# Generamos el archivo de hash para archivo3.txt
print("\n--- Generación de archivo de hash para archivo3.txt ---")
generar_archivo_hash("archivo3.txt")  # Esto crea "archivo3.txt.sha256"

# Podríamos leer los archivos .sha256 para comprobar que los hashes coinciden con los esperados.
# para archivo1.txt.sha256:
with open("archivo1.txt.sha256", "r") as f:
    hash_leido = f.read().strip()
print(f"\nHash guardado de archivo1.txt: {hash_leido}")

with open("archivo2.txt.sha256", "r") as f:
    hash_leido = f.read().strip()
print(f"\nHash guardado de archivo2.txt: {hash_leido}")

with open("archivo3.txt.sha256", "r") as f:
    hash_leido = f.read().strip()
print(f"\nHash guardado de archivo3.txt: {hash_leido}")

```

B. Detección de integridad

```python
import hashlib

# Función para calcular el hash SHA-256 de un archivo
def hash_archivo(ruta):
    sha = hashlib.sha256()  # Creamos un objeto hash vacío
    # Abrimos el archivo en modo lectura de bytes ('rb')
    with open(ruta, 'rb') as f:
        # Leemos el archivo en bloques de 4096 bytes para no cargarlo todo en memoria
        for bloque in iter(lambda: f.read(4096), b""):
            sha.update(bloque)  # Vamos actualizando el hash con cada bloque
    return sha.hexdigest()  # Devolvemos el hash final en formato hexadecimal

# Función para comparar dos archivos usando sus hashes
def comparar_archivos_hash(ruta1, ruta2):
    # Calculamos el hash de cada archivo y vemos si son iguales
    return hash_archivo(ruta1) == hash_archivo(ruta2)

# Creamos tres archivos de texto con contenido
with open("archivo1.txt", "w") as f:
    f.write("El profesor nos coloca cinco uwu")
with open("archivo2.txt", "w") as f:
    f.write("El profesor nos coloca cinco uwu")  # Mismo contenido que archivo1
with open("archivo3.txt", "w") as f:
    f.write("El profesor nos coloca tres unu")   # Contenido diferente

# Comparamos los archivos usando la función
print(f"\n¿archivo1.txt y archivo2.txt iguales? {comparar_archivos_hash('archivo1.txt', 'archivo2.txt')}")
# Debería dar True porque tienen el mismo contenido

print(f"¿archivo1.txt y archivo3.txt iguales? {comparar_archivos_hash('archivo1.txt', 'archivo3.txt')}")
# Debería dar False porque son diferentes

print(f"¿archivo2.txt y archivo3.txt iguales? {comparar_archivos_hash('archivo2.txt', 'archivo3.txt')}")
# También False, por la misma razón
```

C. HMAC

```python
import hashlib   # Necesario para indicar que usaremos SHA-256
import hmac      # Importamos la librería para generar HMAC

# Función para generar un HMAC a partir de un mensaje y una clave
def generar_hmac(mensaje, clave_hmac):
    # hmac.new necesita la clave y el mensaje en bytes, por eso usamos .encode()
    # El tercer parámetro es la función hash que queremos usar (SHA-256)
    # .hexdigest() nos devuelve el resultado en formato hexadecimal
    return hmac.new(clave_hmac.encode(), mensaje.encode(), hashlib.sha256).hexdigest()

# Función para verificar si un HMAC recibido es correcto
def verificar_hmac(mensaje, clave_hmac, hmac_recibido):
    # Calculamos el HMAC que debería tener el mensaje con la misma clave
    calculado = generar_hmac(mensaje, clave_hmac)
    # Comparamos el HMAC calculado con el que nos dieron
    # Usamos compare_digest para hacer la comparación de forma segura
    # (evita que un atacante pueda medir tiempos y adivinar)
    return hmac.compare_digest(calculado, hmac_recibido)

print("\n=== HMAC ===")
msg = "¿Cómo vendió el horticultor sus productos en la red oscura? Utilizó el enrutamiento de cebolla."
# Clave secreta que solo conocen el emisor y el receptor
key_hmac = "Cebollas"
# Generamos el HMAC para ese mensaje
hmac_val = generar_hmac(msg, key_hmac)
print(f"HMAC calculado: {hmac_val}")

# Verificamos que el HMAC es correcto
print(f"Verificación correcta: {verificar_hmac(msg, key_hmac, hmac_val)}")

# Ahora probamos con el mismo mensaje pero modificado (le añadimos una 'X' al final)
# La verificación debería dar False porque el mensaje ya no es el original
print(f"Verificación con mensaje alterado: {verificar_hmac(msg + 'X', key_hmac, hmac_val)}")
```

### Autenticación con hash y sal

```python
import os          # Para generar valores aleatorios (la sal)
import hashlib     # Para usar funciones hash
import hmac        # Para comparar de forma segura

# Este diccionario hace de "base de datos" de usuarios
# Guardará para cada usuario: su hash de contraseña y la sal que se usó
base_usuarios = {}

# Función para registrar un nuevo usuario
def registrar(usuario, password):
    # Generamos una sal aleatoria de 16 bytes (128 bits) y la convertimos a hexadecimal
    sal = os.urandom(16).hex()
    # Calculamos el hash de la contraseña usando SHA-256
    # - 'sha256': algoritmo hash que vamos a usar
    # - password.encode(): convertimos la contraseña a bytes
    # - sal.encode(): convertimos la sal a bytes
    # - 100000: número de iteraciones (con más iteraciones es más seguro, pero más lento)
    hash_pw = hashlib.pbkdf2_hmac('sha256', password.encode(), sal.encode(), 100000).hex()
    # Guardamos el hash y la sal en nuestra "base de datos"
    base_usuarios[usuario] = {"hash": hash_pw, "sal": sal}
    print(f"Usuario '{usuario}' registrado con éxito.")

# Función para iniciar sesión
def login(usuario, password):
    # Si el usuario no existe, devolvemos False
    if usuario not in base_usuarios:
        return False
    # Recuperamos los datos del usuario (hash guardado y sal)
    datos = base_usuarios[usuario]
    # Calculamos el hash de la contraseña que nos acaban de dar, usando la misma sal y el mismo número de iteraciones
    hash_calculado = hashlib.pbkdf2_hmac('sha256', password.encode(), datos["sal"].encode(), 100000).hex()
    # Comparamos el hash recién calculado con el que tenemos guardado
    # Usamos compare_digest para que la comparación sea segura
    return hmac.compare_digest(hash_calculado, datos["hash"])

# Menú interactivo para probar el sistema
while True:
    print("\n--- Menú de Autenticación ---")
    print("1. Registrar nuevo usuario")
    print("2. Iniciar sesión")
    print("3. Salir")
    opcion = input("Seleccione una opción: ").strip()

    if opcion == '1':
        user = input("Ingrese nombre de usuario: ").strip()
        # Verificamos si el usuario ya existe en nuestra base
        if user in base_usuarios:
            print("El usuario ya existe. Intente con otro nombre.")
            continue  # Volvemos al menú principal
        pwd = input("Ingrese contraseña: ").strip()
        registrar(user, pwd)

    elif opcion == '2':
        user = input("Usuario: ").strip()
        pwd = input("Contraseña: ").strip()
        if login(user, pwd):
            print("Inicio de sesión exitoso. ¡Bienvenido!")
        else:
            print("Usuario o contraseña incorrectos.")

    elif opcion == '3':
        print("Saliendo del modo interactivo.")
        break

    else:
        print("Opción no válida.")
```

### Firma Digital

```python
# Importamos las herramientas necesarias de la librería pycryptodome
from Crypto.PublicKey import RSA        # Para generar claves RSA (pública y privada)
from Crypto.Signature import pkcs1_15   # Para firmar y verificar
from Crypto.Hash import SHA256           # Para calcular el hash SHA-256 del mensaje

def firma_digital():
    """
    Función que muestra cómo se genera y verifica una firma digital con RSA.
    """
    print("\n=== Firma Digital RSA ===")

    # 1. Generamos un par de claves RSA de 2048 bits
    key = RSA.generate(2048) # RSA.generate(2048) crea un objeto que contiene la clave privada y la pública
    # Exportamos la clave privada a formato PEM (texto) para poder guardarla después
    private_key = key.export_key()
    # Exportamos la clave pública (se puede obtener de la privada)
    public_key = key.publickey().export_key()

    print("Claves RSA generadas.")

    # 2. Definimos el mensaje que vamos a firmar
    mensaje = "¿Por qué la IA se negó a jugar al escondite? ¡Porque sabía que no podía ocultar sus violaciones de datos!".encode('utf-8')
    # Mostramos el mensaje original (lo decodificamos para que se vea como texto)
    print(f"Mensaje original: {mensaje.decode('utf-8')}")

    # 3. Calculamos el hash SHA-256 del mensaje
    # El hash es como una "huella digital" del mensaje, de tamaño fijo
    h = SHA256.new(mensaje)

    # 4. Firmamos el hash usando la clave privada
    # pkcs1_15.new(key) crea un objeto que sabe firmar
    firmador = pkcs1_15.new(key)
    # El método sign() genera la firma a partir del hash
    firma = firmador.sign(h)
    # Mostramos la firma en hexadecimal (solo los primeros 64 caracteres para que no sea muy larga)
    print(f"Firma generada (hex): {firma.hex()[:64]}... ")

    # 5. Verificamos la firma usando la clave pública
    # Creamos un verificador con la clave pública
    verificador = pkcs1_15.new(key.publickey())
    try:
        # Intentamos verificar: pasamos el hash original y la firma
        verificador.verify(h, firma)
        print("Verificación exitosa: la firma es válida.")
    except (ValueError, TypeError):
        # Si la firma no es válida, se lanza una excepción y entramos aquí
        print("Firma inválida.")

    # 6. Demostramos que si el mensaje cambia, la verificación falla
    mensaje_modificado = "¿Por qué la IA se negó a jugar al escondite? no se".encode('utf-8')
    print(f"Mensaje modificado: {mensaje_modificado.decode('utf-8')}")
    # Calculamos el hash del mensaje modificado
    h_mod = SHA256.new(mensaje_modificado)
    try:
        # Intentamos verificar la misma firma (la original) contra el hash del mensaje modificado
        verificador.verify(h_mod, firma)
        # (no debería pasar porque el mensaje es diferente)
        print("La verificación con mensaje modificado pasa.")
    except (ValueError, TypeError):
        # la verificación falla porque el mensaje fue alterado
        print("Verificación fallida: el mensaje fue alterado.")

    # Guardamos las claves en archivos para poder usarlas en otro momento si queremos
    with open("clave_privada.pem", "wb") as f:
        f.write(private_key)
    with open("clave_publica.pem", "wb") as f:
        f.write(public_key)
    print("\nClaves guardadas en 'clave_privada.pem' y 'clave_publica.pem'.")

# Llamamos a la función para que se ejecute
if __name__ == "__main__":
    firma_digital()
```

### Introducción Práctica a Blockchain

```python
import hashlib
import datetime

class BloqueEstudiante:
    # Bloque que almacena la información de un estudiante.
    def __init__(self, nombre, edad, codigo, genero, hash_anterior=""):
        # Guardamos los datos del estudiante
        self.timestamp = datetime.datetime.now().isoformat()
        self.nombre = nombre
        self.edad = edad
        self.codigo = codigo
        self.genero = genero
        self.hash_anterior = hash_anterior  # Hash del bloque anterior en la cadena
        # Calculamos el hash de este bloque
        self.hash = self.calcular_hash()

    def calcular_hash(self):

        # Calcula el hash SHA-256 concatenando todos los campos del bloque.

        contenido = (str(self.timestamp) + self.nombre + str(self.edad) +
                     self.codigo + self.genero + self.hash_anterior)
         # Aplicamos SHA-256 a ese texto y devolvemos el resultado en hexadecimal
        return hashlib.sha256(contenido.encode()).hexdigest()

    def __repr__(self):
        #Para mostrar el bloque de forma legible.
        return (f"Bloque(\n"
                f"  timestamp    : {self.timestamp}\n"
                f"  nombre       : {self.nombre}\n"
                f"  edad         : {self.edad}\n"
                f"  código       : {self.codigo}\n"
                f"  género       : {self.genero}\n"
                f"  hash_anterior: {self.hash_anterior[:10]}...\n"
                f"  hash         : {self.hash[:10]}...\n)")


class BlockchainEstudiantil:
    # Cadena de bloques simple, sin prueba de trabajo.
    def __init__(self):
        # La cadena comienza con un bloque génesis
        self.cadena = [self.crear_bloque_genesis()]

    def crear_bloque_genesis(self):
        #Crea el primer bloque de la cadena (sin datos reales).
        return BloqueEstudiante(
            nombre="Génesis",
            edad=0,
            codigo="000000",
            genero="N/A",
            hash_anterior="0"  # El primer bloque no tiene anterior
        )

    # Devuelve el último bloque de la cadena
    def obtener_ultimo_bloque(self):
        return self.cadena[-1]

    # Agrega un nuevo estudiante a la cadena (crea un bloque y lo añade)
    def agregar_estudiante(self, nombre, edad, codigo, genero):
       # Creamos un nuevo bloque; el hash anterior es el hash del último bloque actual
        nuevo = BloqueEstudiante(
            nombre, edad, codigo, genero,
            hash_anterior=self.obtener_ultimo_bloque().hash
        )
        # Lo añadimos a la cadena
        self.cadena.append(nuevo)
        print(f"Estudiante registrado: {nombre} (código {codigo})")

    def es_cadena_valida(self):

        #Verifica la integridad de toda la cadena:
          #- Cada bloque tiene el hash correcto (recalculado).
          #- Cada bloque apunta al hash correcto del bloque anterior.

        # Empezamos desde el bloque 1 (el índice 0 es el génesis)
        for i in range(1, len(self.cadena)):
            actual = self.cadena[i]
            anterior = self.cadena[i-1]

            # Recalculamos el hash del bloque actual
            if actual.hash != actual.calcular_hash():
                print(f"ERROR: El hash del bloque de {actual.nombre} no coincide.")
                return False

            # Verificamos que el enlace con el anterior sea correcto
            if actual.hash_anterior != anterior.hash:
                print(f"ERROR: El bloque de {actual.nombre} no apunta correctamente al anterior.")
                return False

        # Si todo está bien, la cadena es válida
        return True

    # Muestra todos los bloques de la cadena
    def mostrar_cadena(self):
        """Imprime todos los bloques de la cadena."""
        print("\n=== CADENA DE BLOQUES ESTUDIANTIL ===\n")
        for i, bloque in enumerate(self.cadena):
            print(f"--- Bloque {i} ---")
            print(bloque)
            print()

# DEMOSTRACIÓN

if __name__ == "__main__":
    # Crear la blockchain
    blockchain = BlockchainEstudiantil()

    # Agregar algunos estudiantes
    print("\n--- Registrando estudiantes en la blockchain ---\n")
    blockchain.agregar_estudiante("Gabriela", 21, "1202419", "F")
    blockchain.agregar_estudiante("Gian Marco", 22, "1202414", "M")
    blockchain.agregar_estudiante("Michelle", 20, "362514", "F")
    blockchain.agregar_estudiante("Maria", 23, "1202412", "F")

    # Mostrar toda la cadena
    blockchain.mostrar_cadena()

    # Verificar integridad
    print("Verificando integridad de la cadena...")
    print("Cadena válida:", blockchain.es_cadena_valida())

    # Demostración de inmutabilidad
    print("\n--- Modificando un bloque (ataque) ---")
    # Accedemos al segundo bloque (índice 1) y cambiamos la edad
    bloque_modificado = blockchain.cadena[1]
    print(f"Bloque original de {bloque_modificado.nombre}: edad={bloque_modificado.edad}")
    bloque_modificado.edad = 99  # Modificamos
    # No recalculamos el hash, así que el hash del bloque ya no corresponde a sus datos
    print(f"Bloque modificado: ahora edad={bloque_modificado.edad} (Hash sin actualizar)")

    # Volvemos a verificar la cadena
    print("\nVerificando cadena después de la modificación...")
    if not blockchain.es_cadena_valida():
        print("La cadena ha sido alterada y ya no es válida.")
    else:
        print("La cadena sigue siendo válida.")
```

### Laboratorio sobre HTTPS

```python
import urllib.request   # Para hacer peticiones HTTP/HTTPS
import urllib.error     # Para manejar errores de las peticiones
import ssl              # Para trabajar con SSL/TLS


def prueba_https():
    print("=== Laboratorio HTTPS ===\n")

    # 1. Hacemos una petición a un sitio que usa HTTPS (conexión segura)
    url_https = "https://www.umng.edu.co/inicio"
    print(f"[HTTPS] Solicitando {url_https} ...")
    try:
        # Añadimos un User-Agent para que el servidor no nos bloquee por ser un script
        req = urllib.request.Request(url_https, headers={'User-Agent': 'Mozilla/5.0'})
        # Hacemos la petición y esperamos respuesta
        with urllib.request.urlopen(req, timeout=5) as respuesta:
            # Mostramos el código de respuesta (200 = OK, 404 = no encontrado, etc.)
            print(f"  Código de respuesta: {respuesta.status} {respuesta.reason}")
            print("  Encabezados relevantes:")
            # Recorremos todos los encabezados de la respuesta
            for header, value in respuesta.getheaders():
                # Solo mostramos algunos encabezados
                if header.lower() in ('content-type', 'server', 'date'):
                    print(f"    {header}: {value}")
            # Leemos los primeros 200 bytes del contenido (para no saturar la pantalla)
            contenido = respuesta.read(200)
            print(f"  Primeros 200 bytes del contenido: {contenido[:100]}...\n")
    except urllib.error.URLError as e:
        # Si hay algún error (red, certificado, etc.), lo mostramos
        print(f"  Error en petición HTTPS: {e}\n")

    # 2. Hacemos una petición a un sitio que usa HTTP (sin cifrar)
    url_http = "http://httpbin.org/"
    print(f"[HTTP] Solicitando {url_http} ...")
    try:
        req = urllib.request.Request(url_http, headers={'User-Agent': 'Mozilla/5.0'})
        with urllib.request.urlopen(req, timeout=5) as respuesta:
            print(f"  Código de respuesta: {respuesta.status} {respuesta.reason}")
            print("  Encabezados relevantes:")
            for header, value in respuesta.getheaders():
                if header.lower() in ('content-type', 'server', 'date'):
                    print(f"    {header}: {value}")
            # Avisamos de que esta conexión no es segura
            print("  (Conexión sin cifrar, no es segura)\n")
    except urllib.error.URLError as e:
        print(f"  Error en petición HTTP: {e}\n")

    # 3. Probamos con un sitio que tiene un certificado expirado (para ver cómo se detecta)
    url_expirado = "https://expired.badssl.com/"
    print(f"Solicitando {url_expirado} ...")
    try:
        # Intentamos abrir la URL (por defecto, Python verifica los certificados)
        with urllib.request.urlopen(url_expirado, timeout=5) as respuesta:
            # Si llegamos aquí, el certificado sería válido
            print(" El certificado es inválido.\n")
    except urllib.error.URLError as e:
        # Comprobamos si el error es por verificación del certificado
        if isinstance(e.reason, ssl.SSLCertVerificationError):
            print("  Error esperado: certificado expirado detectado.")
            print(f"  Detalle: {e.reason}\n")
        else:
            print(f"  Otro error: {e}\n")

# Llamamos a la función
if __name__ == "__main__":
    prueba_https()
```
